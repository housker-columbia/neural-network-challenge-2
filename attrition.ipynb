## Part 1: Preprocessing

```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import numpy as np
from tensorflow.keras.models import Model
from tensorflow.keras import layers

#  Import and read the attrition data
attrition_df = pd.read_csv('https://static.bc-edx.com/ai/ail-v-1-0/m19/lms/datasets/attrition.csv')
attrition_df.head()
```

```python
# Determine the number of unique values in each column.
attrition_df.nunique()
```

```python
# Create y_df with the Attrition and Department columns
y_df = attrition_df[["Attrition", "Department"]]
y_df.head()
```

```python
# Create a list of at least 10 column names to use as X data
# Going with the example for now
features = ["Education", "Age", "DistanceFromHome", "JobSatisfaction", "OverTime", "StockOptionLevel", "WorkLifeBalance", "YearsAtCompany", "YearsSinceLastPromotion", "NumCompaniesWorked"]

# Create X_df using your selected columns
X_df = attrition_df[features]

# Show the data types for X_df
X_df.dtypes
```

```python
# Split the data into training and testing sets
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_df, y_df, random_state=1)
```

```python
# Convert your X data to numeric data types however you see fit
# Add new code cells as necessary

from sklearn.preprocessing import LabelEncoder    # OneHotEncoder could also be used

# Create an instance of the label encoder
le = LabelEncoder()

X_train["OverTime"] = le.fit_transform(X_train["OverTime"])
X_test["OverTime"] = le.transform(X_test["OverTime"])

print(X_train["OverTime"].value_counts())
print(X_test["OverTime"].value_counts())

print(X_train.dtypes)
print(X_test.dtypes)
```

```python
X_train
```

```python
# Create a StandardScaler
ss = StandardScaler()

# Fit the StandardScaler to the training data
X_scaler = ss.fit(X_train)

# Scale the training and testing data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
pd.DataFrame(X_train_scaled, columns=X_train.columns).head()
```

```python
y_df["Department"].value_counts()
```

```python
# Create a OneHotEncoder for the Department column
from sklearn.preprocessing import OneHotEncoder

ohed = OneHotEncoder(sparse_output=False, handle_unknown='ignore').set_output(transform="pandas")

# Fit the encoder to the training data
ohed_encoder = ohed.fit(y_train[["Department"]])

# Create two new variables by applying the encoder
# to the training and testing data
# y_train_ohe = pd.concat([y_train.drop(columns=["Department"]), ohed_encoder.transform(y_train[["Department"]])], axis=1)
# y_test_ohe = pd.concat([y_test.drop(columns=["Department"]), ohed_encoder.transform(y_test[["Department"]])], axis=1)
y_train_ohed = ohed_encoder.transform(y_train[["Department"]])
y_test_ohed = ohed_encoder.transform(y_test[["Department"]])
display(y_train_ohed)
```

```python
# Create a OneHotEncoder for the Attrition column
ohea = OneHotEncoder(sparse_output=False, handle_unknown='ignore').set_output(transform="pandas")

# Fit the encoder to the training data
ohea_encoder = ohea.fit(y_train[["Attrition"]])

# Create two new variables by applying the encoder
# to the training and testing data
# y_train_ohe = pd.concat([y_train_ohe.drop(columns=["Attrition"]), ohea_encoder.transform(y_train[["Attrition"]])], axis=1)
# y_test_ohe = pd.concat([y_test_ohe.drop(columns=["Attrition"]), ohea_encoder.transform(y_test[["Attrition"]])], axis=1)
y_train_ohea = ohea_encoder.transform(y_train[["Attrition"]])
y_test_ohea = ohea_encoder.transform(y_test[["Attrition"]])
display(y_train_ohea)
display(y_test_ohea)
```

## Create, Compile, and Train the Model

```python
# Find the number of columns in the X training data
# len(X_train.columns)
ncol = X_train_scaled.shape[1]

# Create the input layer
input_layer = layers.Input(shape=(ncol,), name='input_features')

# Create at least two shared layers
shared_layer1 = layers.Dense(64, activation='relu')(input_layer)
shared_layer2 = layers.Dense(32, activation='relu')(shared_layer1)
```

```python
# Create a branch for Department
# with a hidden layer and an output layer

# Create the hidden layer
dept_dense = layers.Dense(64, activation='relu')(shared_layer2)

# Create the output layer
dept_output = layers.Dense(len(y_df["Department"].unique()),
                             activation='softmax',
                             name='dept_output')(dept_dense)
```

```python
# Create a branch for Attrition
# with a hidden layer and an output layer

# Create the hidden layer
att_dense = layers.Dense(64, activation='relu')(shared_layer2)

# Create the output layer
att_output = layers.Dense(len(y_df["Attrition"].unique()),
                             activation='sigmoid',
                             name='att_output')(att_dense)

```

```python
# Create the model
model = Model(inputs=input_layer, outputs=[
    dept_output,
    att_output
])

# Compile the model
model.compile(optimizer='adam',
              loss={'dept_output': 'categorical_crossentropy',
                    'att_output': 'binary_crossentropy'},
              metrics={'dept_output': 'accuracy',
                       'att_output': 'accuracy'})

# Summarize the model
model.summary()
```

```python
# Train the model
model.fit(
    X_train_scaled,
    {
        'dept_output': y_train_ohed,
        'att_output': y_train_ohea
    },
    epochs=10,
    batch_size=32
)
```

```python
# Evaluate the model with the testing data
results = model.evaluate(np.array(X_test), {
        'dept_output': y_test_ohed,
        'att_output': y_test_ohea
    })
```

```python
# Print the accuracy for both department and attrition
# Results contains the following in order:
# Total loss (sum of all losses)
# Loss for dept_output
# Loss for att_output
# Accuracy for dept_output
# Accuracy for att_output

print(f"Department predictions accuracy: {results[3]}")
print(f"Attrition predictions accuracy: {results[4]}")
```
